```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
import datetime

nasa = pd.read_csv("./data/nasa_facebook_statuses.csv", 
                   usecols=["status_published","num_reactions",
                            "num_comments","num_shares","num_likes"])

nasa["status_published"] = pd.to_datetime(nasa["status_published"])

filter =(nasa['status_published'] >= '2016-5-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]

nasa['day'] = nasa.apply(lambda row: row["status_published"].day,axis=1)

nasa_by_day = nasa.groupby(by='day').sum()
nasa_by_day
```

```python
nasa_by_day[['num_shares','num_comments']].plot()
```

```python
nasa_by_day.plot(y=['num_reactions','num_likes'])
```

```python
nasa_by_day.plot(figsize=(15,5))
```

```python
nasa_by_day.plot(figsize=(15,5),use_index=False,title='NASA Fb Activity',grid=True)
```

```python
nasa_by_day.plot(figsize=(15,10),subplots=True,grid=True)
```

```python
nasa_by_day.plot(figsize=(15,5),grid=True,logy=True)
```

```python
nasa_by_day.plot(figsize=(15,5),grid=True,xticks=range(1,32),rot=30)
```

```python
nasa_by_day[:10].plot(figsize=(15,5),grid=True,xticks=range(1,11),table=True)
```
