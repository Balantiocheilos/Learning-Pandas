```python
import pandas as pd
import numpy as np
```

```python
fortune=pd.read_csv('./data/Fortune_500_2017.csv',index_col='Rank',
                   usecols=['Rank','Title','Employees','Profits','Assets'])
```

```python
fortune['RankByEmployees']=fortune['Employees'].rank(ascending=False)
fortune.head()
```

```python
fortune['RankByProfits']=fortune['Profits'].rank(ascending=False)
fortune.head()
```

```python
isEmployeesRankFirst10 = fortune['RankByEmployees'] <= 10
isProfitRankFirst10 = fortune['RankByProfits'] <= 10
fortune['ProfitPerEmployee'] = fortune['Profits'] * 1000000 / fortune['Employees']
fortune.head()
```

```python
fortune[isEmployeesRankFirst10 | isProfitRankFirst10]
```

```python
isEmployeesRankMore400 = fortune.RankByEmployees >= 400
fortune[isEmployeesRankMore400 & isProfitRankFirst10]
```

```python
fortune.sort_values(by='ProfitPerEmployee',ascending=False).head(20)
```
