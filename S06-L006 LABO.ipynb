```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./data/marathon_results_2016.csv', index_col='Bib',
 usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
df['Pace'] = df['Pace'].replace('-', 0)
 
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['Pace'] = df['Pace'].apply(pd.to_timedelta)
 
'''df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))

df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)'''

df.head()
```

```python
group_city=df.groupby('City')
```

```python
group_city.head()
```

```python
for cities,cities_data in group_city:
    print(cities,
         cities_data['40K'].min(),
          len(cities),
         )
```

```python
cities_champions=pd.DataFrame()
for cities,cities_data in group_city:
    cities_champions=cities_champions.append(cities_data.nsmallest(1,'40K'))
```

```python
group_countries=df.groupby('Country')
best_pace=pd.DataFrame()
for countries,countries_data in group_countries:
    best_pace=best_pace.append(countries_data.nsmallest(1,'Pace'))
best_pace.sort_values('Pace',inplace=True)
```
