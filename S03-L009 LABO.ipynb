```python
import pandas as pd
import numpy as np
fortune=pd.read_csv('./data/Fortune_500_2017.csv',index_col='Rank',
                   usecols=['Rank','Title','Employees','Profits','Assets'])
```

```python
fortune['Employees'].rank(ascending=False).head()
```

```python
fortune['RankByEmployee']=fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune.nsmallest(n=10,columns='RankByEmployee')
```

```python
fortune['RankByProfits']=fortune['Profits'].rank(ascending=False)
```

```python
fortune.nsmallest(n=10,columns='RankByProfits')
```
