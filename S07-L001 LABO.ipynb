```python
mrbean=pd.read_csv('./data/mrbean_facebook_statuses_join.csv',
                  usecols=['status_message_mrbean','status_type','num_likes'])
```

```python
nasa=pd.read_csv('./data/nasa_facebook_statuses_join.csv',
                usecols=['status_message_nasa','status_type','num_likes'])
```
